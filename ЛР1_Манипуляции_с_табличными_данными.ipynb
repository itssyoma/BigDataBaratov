## Задание 6

1. Для каждого типа транзакций рассчитайте максимальную сумму прихода на карту (из строго положительных сумм по столбцу `amount`) отдельно для мужчин и женщин (назовите ее `max_income`). Оставьте по 10 типов транзакций для мужчин и для женщин, наименьших среди всех типов транзакций по полученным значениям `max_income`.
2. Среди типов транзакций, найденных в пункте 1, выделите те, которые встречаются одновременно и у мужчин, и у женщин.

````python
df = transactions_merged.copy()
df_g = df.dropna(subset=['gender']).copy()

# Только приходы
pos = df_g[df_g['amount'] > 0].copy()

# max_income по (gender, tr_type)
max_income = (
    pos.groupby(['gender', 'tr_type'])['amount']
       .max()
       .reset_index(name='max_income')
)

# 10 наименьших типов по max_income для каждого пола
bottom10_by_gender = (
    max_income.sort_values(['gender', 'max_income'], ascending=[True, True])
              .groupby('gender')
              .head(10)
)

# Добавим описание типа
bottom10_by_gender = bottom10_by_gender.merge(
    df_g[['tr_type', 'tr_description']].drop_duplicates('tr_type'),
    on='tr_type',
    how='left'
)

print("10 типов транзакций с наименьшим max_income отдельно по полу:")
print(bottom10_by_gender.sort_values(['gender', 'max_income']))

# Типы, которые встречаются и у мужчин, и у женщин среди найденных bottom-10
sets = bottom10_by_gender.groupby('gender')['tr_type'].apply(set).to_dict()

common_types = set.intersection(*sets.values()) if len(sets) >= 2 else set()
common_df = bottom10_by_gender[bottom10_by_gender['tr_type'].isin(common_types)] \
            .sort_values(['tr_type', 'gender'])

print("\nТипы, которые встречаются одновременно у обоих полов (в этих bottom-10 списках):")
print(common_df if not common_df.empty else "Пересечений нет.")
````

## Задание 7

1. Найдите суммы затрат по каждой категории (`mcc`) для мужчин и для женщин.
2. Найдите топ-10 категорий с самыми большими относительными модулями разности в тратах для разных полов (в ответе должны присутствовать описания mcc-кодов).

````python
df = transactions_merged.copy()
df_g = df.dropna(subset=['gender']).copy()

# Только траты (отрицательные)
sp = df_g[df_g['amount'] < 0].copy()

# Сумма трат по mcc и полу (сумма будет отрицательная — оставляем как есть)
mcc_spend = (
    sp.groupby(['mcc_code', 'mcc_description', 'gender'])['amount']
      .sum()
      .reset_index()
)

# Пивот: отдельно суммы для полов
pivot = mcc_spend.pivot_table(
    index=['mcc_code', 'mcc_description'],
    columns='gender',
    values='amount',
    aggfunc='sum',
    fill_value=0
)

# Убедимся, что есть ровно две колонки полов (обычно 0 и 1)
cols = list(pivot.columns)
if len(cols) < 2:
    print("Недостаточно групп по полу для сравнения (нужны минимум 2 значения gender).")
else:
    g0, g1 = cols[0], cols[1]
    a = pivot[g0]
    b = pivot[g1]

    # Относительный модуль разности:
    # abs(a-b) / (abs(a)+abs(b))  (устойчиво, не зависит от знака, 0..1)
    denom = (a.abs() + b.abs()).replace(0, np.nan)
    pivot['rel_abs_diff'] = (a - b).abs() / denom

    top10 = pivot.sort_values('rel_abs_diff', ascending=False).head(10).reset_index()
    top10 = top10[['mcc_code', 'mcc_description', g0, g1, 'rel_abs_diff']]

    print("Топ-10 MCC с наибольшими относительными модулями разности трат по полу:")
    print(top10)
````

## Задание 8

1. Из поля `tr_datetime` выделите час `tr_hour`, в который произошла транзакция, как первые 2 цифры до `":"`.
2. Посчитайте количество транзакций со значением в столбце `amount` строго меньше `0` в ночное время для мужчин и женщин. Ночное время – это промежуток с 00:00 по 06:00 часов.

````python
df = transactions_merged.copy()
df_g = df.dropna(subset=['gender']).copy()

# tr_datetime выглядит как "88 23:36:53" -> берем часть после пробела, затем первые 2 цифры до ":"
time_part = df_g['tr_datetime'].astype(str).str.split(' ', n=1).str[1]
df_g['tr_hour'] = pd.to_numeric(time_part.str.slice(0, 2), errors='coerce')

# Ночь: 00:00–06:00 (часы 0..6 включительно)
night = df_g[(df_g['amount'] < 0) & (df_g['tr_hour'].between(0, 6, inclusive='both'))]

cnt = night.groupby('gender').size().reset_index(name='night_negative_tr_count')

print("Количество ночных трат (amount < 0) по полу (00:00–06:00):")
print(cnt)
````

````python
df = transactions_merged.copy()

time_part = df['tr_datetime'].astype(str).str.split(' ', n=1).str[1]
df['tr_hour'] = pd.to_numeric(time_part.str.slice(0, 2), errors='coerce')

# Ночные траты: amount < 0 и 00:00–06:00
night_spend = df[(df['amount'] < 0) & (df['tr_hour'].between(0, 6, inclusive='both'))].copy()

# Группировка по категории MCC: считаем кол-во трат и средний чек (сумма будет отрицательная)
mcc_night_stats = (
    night_spend.groupby(['mcc_code', 'mcc_description'])['amount']
    .agg(night_spend_count='size', mean_night_spend='mean', total_night_spend='sum')
    .reset_index()
)

# Категория, где БОЛЬШЕ ВСЕГО трат (по количеству транзакций)
top_mcc = mcc_night_stats.sort_values('night_spend_count', ascending=False).head(1)

print("Категория с наибольшим числом ночных трат (00:00–06:00):")
display(top_mcc)
````
